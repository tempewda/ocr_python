## 00: loading the necessary packages

```python
import pytesseract
from PIL import Image
```

## 01: setting the paths

```python
img_file = "data/page_03.jpg"
no_noise = "temp/no_noise.jpg"
```

## 02: display the image

```python
img = Image.open(no_noise)
display(img)
```

## 03: run OCR using Pytesseract

```python
ocr_result = pytesseract.image_to_string(img)
```

## 04: output

```python
# Store the OCR result in a text file
output_file = "temp/ocr_result.txt"
with open(output_file, "w") as file:
    file.write(ocr_result)

# Display the output
print (ocr_result)

print("\n\n\nOCR result saved to:", output_file)
```
