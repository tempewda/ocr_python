### displaying the image on collab

```python
from google.colab.patches import cv2_imshow
from skimage import io

# display the image in collab
def im_show(im_path):
  image = io.imread(im_path)
  cv2_imshow(image)
```

```python
im_show(im_path)
```

### working with PIL

```python
from PIL import Image
im_path = "ocr_python/data/page_01.jpg"
im = Image.open(im_path)
```

```python
print(im.format, im.size, im.mode)
```

```python
im.save("ocr_python/temp/page_01_copy.jpg")
im_copy = Image.open("ocr_python/temp/page_01_copy.jpg")

print(im_copy.format, im_copy.size, im_copy.mode)
```
